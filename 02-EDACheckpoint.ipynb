## Background and Prior Work

The RMS Titanic was a British ocean liner infamous for its sinking after colliding with an iceberg on April 15, 1912, resulting in ~1500 people out of the 2200+ passengers dying from the wreck. The data set we are analyzing includes information on 891 passengers, including their passenger class, age, sex, and their accompaniment. Even though considered a catastrophic event, which are usually described as random or indiscriminate, research consistently shows that social factors shape survival outcomes. Disaster research reveals that survival is influenced by social factors, such as socioeconomic status (SES), age, and gender, rather than pure chance. 

Similar studies:

A paper from Bowdoin College highlights that the Titanic, like many disasters, had "differential effects" in which survival and suffering were unequally distributed across class, age, and gender lines. It emphasized that first-class passengers had a survival rate of about 62%, compared to 41% in second class and only 25% in third class (Bowdoin), showcasing stark SES differences and translation to better survival hood. The aper theorizes that the higher status translates into better access to information, lifeboats, and life-saving opportunities.  The paper also incorporates a diagram of the Titanic, showcasing that third class passengers were located lower and at the ends of the boat, second class lower middle, and first class upper middle. This diagram allows for the inference that if the boat were sinking, the first-class passengers had more time to reach life-saving measures. The same piece showcases that women and children had much higher rates of survival, with 75 and 50 percent, respectively, while men only had about 20% survival. This follows suit with the order Captain E.J. Smith gave to officers, "women and children first." The paper also mentions that certain officers interpreted these orders differently, with Officer Lightroller taking the order to mean women and children only, lowering lifeboats even before full capacity to prioritize women and children's survival.

In a more formal analysis, Frey, Heggli, and Osterloh used individual-level Titanic data to test how economic resources, age, and gender affected the probability of survival. Aside from reconfirming that class and gender played a large role in the probability of survival, they showed age mattered as well. People in their "prime" (ages around 15-35) were more likely to survive than older passengers (Frey et al.), suggesting that physical mobility and stamina interacted with the other categories to shape outcomes.

A study similar to our proposed topic, but with a slightly different data set, concluded that gender was, in fact, the key determining factor in survivability.1 Because this is a historical topic, we can safely assume that there will be no issues with outdated data or analysis, but the data sets being analyzed may have varying results. Despite being the same size, with 891 passengers, the data set we are analyzing has some additional information that may reveal more nuance in the death statistics. Additionally, the individuals may be different, meaning a different population of the Titanic survivors was sampled.

An article from the NIH states that, despite the widespread belief of prioritizing women and children in maritime accidents, women in fact have a significant survival disadvantage in these situations 2. It further asserts that despite the survival rate of women on the Titanic being significantly higher, in the grander scheme of maritime accidents reflects the opposite.

#### Section 2 of EDA - If and how passenger class had a significant correlation with survival rates

To find relationship between passenger class had a significant correlation with surivival rates, which implies possible correlation with socio-economic class, we will be performing EDA on passenger class (pclass [1, 2, 3]) and survival rate (survived [0, 1])

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/02-processed/titanic_clean.csv')

print(df['survived'].value_counts())
print(df['pclass'].value_counts().sort_index())

```

The dataset holds 714 passengers — a clean subset of the original 891-row dataset with incomplete records removed. Overall, only 40.6% survived. The class breakdown is notably uneven: Third Class passengers account for 49.7% of all passengers (355 of 714), nearly as many as First and Second combined. This is important context, as any aggregate count will be naturally dominated by Third Class, so we should work in rates rather than raw counts. The deck column is missing for 96% of passengers and won't be useful for this analysis. Additionally, decks were shared between the classes, so we will be omitting deck level from this analysis completely.

##### Survival Rate by Class — Counts and Bar Plot

```python
survival_by_class = df.groupby('pclass')['survived'].agg(['mean', 'sum', 'count'])
survival_by_class.columns = ['survival_rate', 'survivors', 'total']
survival_by_class['survival_rate'] = survival_by_class['survival_rate'].round(3)
print(survival_by_class)

survival_by_class['survival_rate'].plot(kind='bar', color=['gold', 'silver', '#cd7f32'])
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

First, we'll look at the raw rates for survival rate per class, in percentages and in a bar plot. With this, we can immediately see that First class has both a higher survival rate and around 40 more total survivors than Second and Third class each, while having close to the same number of passengers as Second class and roughly half as many passengers as Third class. First Class passengers survived at a rate of 65.6%, which is nearly nearly 2.7× higher than Third Class passengers at 23.9%.

This striking difference shows how class may have significantly shaped survival odds.

##### Crosstab Heatmap of Survival Data

```python
ct = pd.crosstab(df['pclass'], df['survived'], normalize='index')
sns.heatmap(ct, annot=True, fmt='.2%', cmap='RdYlGn')
plt.title('Survival Rate by Class (% of each class)')
plt.show()
```

Next, we'll use a heatmap to see how the ratio of survival was affected in each class. Second class stands out as the most even split, with roughly half surviving and half not. First and Third class have inverted patterns, with Third class being the most dramatic split.

##### Is the Correlation Statistically Significant?

```python
from scipy.stats import chi2_contingency
import numpy as np

ct_raw = pd.crosstab(df['pclass'], df['survived'])
chi2, p, dof, expected = chi2_contingency(ct_raw)

n = ct_raw.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(ct_raw.shape) - 1)))

print(f"Chi2:      {chi2:.2f}")
print(f"P-value:   {p:.20f}")
```

Because both pclass and survived are categorical variables, we'll use the Chi-Square test of independence to look for significance. The result is extremely unambiguous, with Chi^2 = 92.90 and p = 0.000, proving that the probability of seeing these relationships between Class and Survival being essentially zero. This EDA strongly supports the conclusion that passenger class was a significant and meaningful predictor of survival on the Titanic. However, class may not be an isolated factor, as ticket prices may have varied between where they were purchased.
