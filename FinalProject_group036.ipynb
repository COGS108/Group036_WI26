## Background and Prior Work

The RMS Titanic was a British ocean liner infamous for its sinking after colliding with an iceberg on April 15, 1912, resulting in ~1500 people out of the 2200+ passengers dying from the wreck. The data set we are analyzing includes information on 891 passengers, including their passenger class, age, sex, and their accompaniment. Even though considered a catastrophic event, which are usually described as random or indiscriminate, research consistently shows that social factors shape survival outcomes. Disaster research reveals that survival is influenced by social factors, such as socioeconomic status (SES), age, and gender, rather than pure chance. 

Similar studies:

A paper from Bowdoin College highlights that the Titanic, like many disasters, had "differential effects" in which survival and suffering were unequally distributed across class, age, and gender lines. It emphasized that first-class passengers had a survival rate of about 62%, compared to 41% in second class and only 25% in third class (Bowdoin), showcasing stark SES differences and translation to better survival hood. The aper theorizes that the higher status translates into better access to information, lifeboats, and life-saving opportunities.  The paper also incorporates a diagram of the Titanic, showcasing that third class passengers were located lower and at the ends of the boat, second class lower middle, and first class upper middle. This diagram allows for the inference that if the boat were sinking, the first-class passengers had more time to reach life-saving measures. The same piece showcases that women and children had much higher rates of survival, with 75 and 50 percent, respectively, while men only had about 20% survival. This follows suit with the order Captain E.J. Smith gave to officers, "women and children first." The paper also mentions that certain officers interpreted these orders differently, with Officer Lightroller taking the order to mean women and children only, lowering lifeboats even before full capacity to prioritize women and children's survival.

In a more formal analysis, Frey, Heggli, and Osterloh used individual-level Titanic data to test how economic resources, age, and gender affected the probability of survival. Aside from reconfirming that class and gender played a large role in the probability of survival, they showed age mattered as well. People in their "prime" (ages around 15-35) were more likely to survive than older passengers (Frey et al.), suggesting that physical mobility and stamina interacted with the other categories to shape outcomes.

#### Teddy - Survivor Age

This section seperates the data into multiple categories, survivors, nonsurvivors, female survivors, male survivors, female nonsurvivors and male non survivors. Using that data we show their size and mean age, then we plot it.

```python
titanic = pd.read_csv('data/02-processed/titanic_clean.csv')

# Has a lot of nil values and we don't need it
titanic = titanic.drop(['deck'], axis=1)
titanic = titanic.dropna(subset=['age'])

nonsurvivors, survivors = [x for _, x  in titanic.groupby(titanic['survived'] == 1)]

female_survivors, male_survivors = [x for _, x  in survivors.groupby(titanic['sex'] == 'male')]
female_nonsurvivors, male_nonsurvivors = [x for _, x  in nonsurvivors.groupby(titanic['sex'] == 'male')]

# Print all the data
print("")
print("Total Passengers (With Valid Age): ", titanic.shape[0])
print("")
print("Num Male Survivors: ", male_survivors.shape[0])
print("Num Female Survivors: ", female_survivors.shape[0])
print("Num Male Nonsurvivors: ", male_nonsurvivors.shape[0])
print("Num Female Nonsurvivors: ", female_nonsurvivors.shape[0])
print("")
print("Average Male Survivors Age: ", male_survivors['age'].mean().round(2))
print("Average Female Survivors Age: ", female_survivors['age'].mean().round(2))
print("Average Male Nonsurvivors Age: ", male_nonsurvivors['age'].mean().round(2))
print("Average Female Nonsurvivors Age: ", female_nonsurvivors['age'].mean().round(2))
print("")
print("Survivor Stats: ", survivors[["age"]].describe().round(2))
print("")
print("Nonsurvivor Stats: ", nonsurvivors[["age"]].describe().round(2))
print("")

# Box plot of age, sex and whether the person survied or not
survivor_plot = sns.boxplot(data=titanic, x='age', y="alive", hue='sex')
survivor_plot.set_title("Survivor vs Nonsurvivor data")
```

We do a Chi-Squared Test to see if there was a relationship between male and female survivalship. We get a very low p-value which tells us to reject the null hypothesis which is to say there is a relationship between gender and survivalship.

Then we do a t-test to see if age influenced survivalship. And from the p-value we can say age influences survivalship.

```python
# Chi squared test to test relationship between male and female survivalship
chi_squared_test = stats.chi2_contingency(
    [[male_survivors.shape[0], male_nonsurvivors.shape[0]], [female_survivors.shape[0], female_nonsurvivors.shape[0]]]
)
print("\nChi-Squared Test:")
print("Statistic: ", chi_squared_test.statistic)
print("P-Value: ", chi_squared_test.pvalue)

# T-Test to test the relationship between age and survivalship
ttest = stats.ttest_ind(survivors["age"], nonsurvivors["age"])
print("\nT-Test:")
print("Statistic: ", ttest.statistic)
print("P-Value: ", ttest.pvalue)
```

#### Harvey - Alone vs Not Alone

In this section, we directly look at whether passengers traveling alone had different survival outcomes  than passengers traveling with family/spouse. We can summarize the survival rates and raw counts by the dataset's Alone indicator, then visualize both the rate difference and composition of outcomes for survived vs not survived. This allows us to establish the baseline relationship we can later check for pairing it with class, sex, and age.

```python
# Always make sure figures directory exists BEFORE saving
os.makedirs("figures", exist_ok=True)

# Load processed data (do NOT reload raw data here)
df = pd.read_csv("data/02-processed/titanic_clean.csv")

# Quick sanity checks
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# Ensure survived is 0/1
df["survived"] = df["survived"].astype(int)

# Ensure alone is boolean (handles True/False, 0/1, strings)
df["alone"] = df["alone"].astype(str).str.lower().map({"true": True, "false": False, "1": True, "0": False})
df["alone"] = df["alone"].fillna(False).astype(bool)

df[["survived", "alone"]].head()
```

1: load + setup + sanity checks

This block loads the processed dataset so our EDA is reproducible. It also ensured that survived is stored as a 0/1 integer and that alone is consistently interpreted as a boolean. These checks prevent plotting and groupby errors caused by mixed datatypes and can help ensure our later caalculations are valid.

```python
survival_by_alone = (
    df.groupby("alone")["survived"]
      .agg(total="count", survived="sum", survival_rate="mean")
      .reset_index()
)

survival_by_alone["label"] = survival_by_alone["alone"].map({True: "Alone", False: "Not Alone"})
survival_by_alone = survival_by_alone.sort_values("alone")  # Not Alone first

# Add useful effect sizes
rate_not = survival_by_alone.loc[survival_by_alone["alone"] == False, "survival_rate"].iloc[0]
rate_alone = survival_by_alone.loc[survival_by_alone["alone"] == True, "survival_rate"].iloc[0]
survival_by_alone["survival_rate"] = survival_by_alone["survival_rate"].round(4)

print("Survival summary (alone vs not alone):")
display(survival_by_alone[["label", "total", "survived", "survival_rate"]])

print(f"Risk difference (Not Alone - Alone): {rate_not - rate_alone:.3f}")
print(f"Relative risk (Not Alone / Alone): {(rate_not / rate_alone) if rate_alone > 0 else np.nan:.3f}")
```

**Summary table: survival by alone status**

Grouping by the `alone` indicator reveals a clear survival gap:

| Group | Passengers | Survivors | Survival Rate |
|---|---|---|---|
| Not Alone | 310 | 160 | 51.6% |
| Alone | 404 | 130 | 32.2% |

Passengers traveling with others survived at 1.6x the rate of those traveling alone (relative risk = 1.60). The absolute risk difference is roughly 19 percentage points.

```python
fig, ax = plt.subplots(figsize=(7, 5))

labels = survival_by_alone["label"].tolist()
rates = survival_by_alone["survival_rate"].tolist()
totals = survival_by_alone["total"].tolist()

bars = ax.bar(labels, rates, edgecolor="white")

ax.set_ylim(0, 1)
ax.set_ylabel("Survival Rate")
ax.set_title("Titanic Survival Rate: Alone vs Not Alone")
ax.spines[["top", "right"]].set_visible(False)

for bar, rate, n in zip(bars, rates, totals):
    ax.text(bar.get_x() + bar.get_width()/2, rate + 0.02,
            f"{rate:.1%}\n(n={n})", ha="center", va="bottom", fontsize=11)

plt.tight_layout()
plt.savefig("figures/eda_alone_survival_rate.png", dpi=150, bbox_inches="tight")
plt.show()
```

**Bar chart: survival rate by alone status**

The bar chart confirms the rate difference visually, with each bar annotated with the survival percentage and sample size. Not-Alone passengers show a markedly higher survival probability (~52%) compared to those traveling alone (~32%). However, because 57% of passengers in the clean dataset traveled alone, this pattern affects a large portion of the population.

```python
fig, ax = plt.subplots(figsize=(7, 5))

survived_counts = survival_by_alone["survived"].tolist()
not_survived_counts = (survival_by_alone["total"] - survival_by_alone["survived"]).tolist()
x = np.arange(len(labels))

ax.bar(x, survived_counts, label="Survived", edgecolor="white")
ax.bar(x, not_survived_counts, bottom=survived_counts, label="Did Not Survive", edgecolor="white")

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel("Number of Passengers")
ax.set_title("Passenger Counts by Survival and Alone Status")
ax.legend()
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("figures/eda_alone_counts_stacked.png", dpi=150, bbox_inches="tight")
plt.show()
```

**Stacked bar chart: outcome composition by alone status**

While the bar chart above focuses on probability, the stacked chart highlights raw outcome counts. Among the 310 passengers traveling with others, 160 survived and 150 did not -- nearly even. Among the 404 alone passengers, only 130 survived versus 274 who did not -- a much more lopsided split. This composition view makes clear that the survival disadvantage for alone passengers is not an artifact of small sample size.

**Section 2: Confounding & Subgroup Patterns (Sex, Class, Age)**

The relationship between traveling alone and survival may be confounded by other variables such as sex, passenger class, and age. For example, if passengers traveling alone skewed toward adult males or third-class travelers -- groups already at higher mortality risk -- the observed survival gap might reflect those structural disadvantages rather than 'alone status' itself. This section examines whether alone status is unevenly distributed across key demographic and socioeconomic groups.

```python
df = pd.read_csv("data/02-processed/titanic_clean.csv")
df["survived"] = df["survived"].astype(int)
df["alone"] = df["alone"].astype(str).str.lower().map({"true": True, "false": False, "1": True, "0": False})
df["alone"] = df["alone"].fillna(False).astype(bool)

sex_alone = (
    df.groupby(["sex", "alone"])["survived"]
      .agg(total="count", survival_rate="mean")
      .reset_index()
)

sex_alone["alone_label"] = sex_alone["alone"].map({True: "Alone", False: "Not Alone"})

print("Survival rate by sex and alone status:")
display(sex_alone.sort_values(["sex", "alone"]))

# Pivot for safe plotting (won't crash if a cell is missing)
sex_piv = sex_alone.pivot(index="sex", columns="alone_label", values="survival_rate").fillna(0)

fig, ax = plt.subplots(figsize=(7, 5))
sex_piv.plot(kind="bar", ax=ax, edgecolor="white")

ax.set_ylim(0, 1)
ax.set_ylabel("Survival Rate")
ax.set_title("Survival Rate by Sex, split by Alone Status")
ax.legend(title="")
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("figures/eda_alone_by_sex.png", dpi=150, bbox_inches="tight")
plt.show()
```

**Survival rate by sex and alone status**

Breaking down survival by sex reveals an important pattern: females traveling alone still survived at 79%, actually slightly higher than females traveling with others (73%). For males, the alone penalty is much larger -- lone males survived at only 17% vs. 28% for males with companions. This suggests that the overall alone-survival gap is driven almost entirely by male passengers, and that 'alone status' interacts strongly with sex.

```python
class_alone = (
    df.groupby(["pclass", "alone"])["survived"]
      .agg(total="count", survival_rate="mean")
      .reset_index()
)

class_alone["alone_label"] = class_alone["alone"].map({True: "Alone", False: "Not Alone"})

print("Survival rate by class and alone status:")
display(class_alone.sort_values(["pclass", "alone"]))

class_piv = class_alone.pivot(index="pclass", columns="alone_label", values="survival_rate").fillna(0)

fig, ax = plt.subplots(figsize=(7, 5))
class_piv.loc[[1,2,3]].plot(kind="bar", ax=ax, edgecolor="white")

ax.set_ylim(0, 1)
ax.set_ylabel("Survival Rate")
ax.set_title("Survival Rate by Passenger Class, split by Alone Status")
ax.set_xticklabels(["1st", "2nd", "3rd"], rotation=0)
ax.legend(title="")
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("figures/eda_alone_by_class.png", dpi=150, bbox_inches="tight")
plt.show()
```

##### Is the Correlation Statistically Significant?

```python
ct_raw = pd.crosstab(df['pclass'], df['survived'])
chi2, p, dof, expected = chi2_contingency(ct_raw)

n = ct_raw.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(ct_raw.shape) - 1)))

print(f"Chi2:      {chi2:.2f}")
print(f"P-value:   {p:.20f}")
```

Because both pclass and survived are categorical variables, we'll use the Chi-Square test of independence to look for significance. The result is extremely unambiguous, with Chi^2 = 92.90 and p = 0.000, proving that the probability of seeing these relationships between Class and Survival being essentially zero. This EDA strongly supports the conclusion that passenger class was a significant and meaningful predictor of survival on the Titanic. However, class may not be an isolated factor, as ticket prices may have varied between where they were purchased.

#### Section 2 of EDA - Preliminary Model Testing

With individual features explored in Section 1, we now take a model-level view. A logistic regression trained on all available features provides a baseline measure of predictive power and reveals which passenger characteristics most strongly drove survival in a joint model.

#### Connor - Logistic Regression of All Features

This section trains a logistic regression model on the full cleaned feature set to produce a baseline survival classifier. The model is evaluated using accuracy and ROC-AUC, and a confusion matrix is used to examine false-positive and false-negative rates. Standardized feature coefficients expose which variables most strongly influenced predictions, and side-by-side bar charts visualize survival rates across the three key demographic groups (sex, class, alone status) to corroborate the model's findings.

```python
X = df[main_features]
y = df['survived']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)

lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
y_pred  = lr.predict(X_test)
y_proba = lr.predict_proba(X_test)[:, 1]
```

**Logistic regression results: accuracy, ROC-AUC, and feature importance**

The model achieves an **accuracy of 80.4%** and a **ROC-AUC of 0.845** on the held-out test set -- a strong baseline. The confusion matrix shows balanced precision and recall for both classes:

| Class | Precision | Recall | F1 |
|---|---|---|---|
| Did Not Survive | 0.84 | 0.84 | 0.84 |
| Survived | 0.76 | 0.76 | 0.76 |

The standardized feature coefficients confirm what Section 1 suggested: **sex** is the dominant predictor of survival, followed by **passenger class** and **fare**. Being male, traveling in a lower class, or paying a lower fare were all associated with significantly reduced survival odds in the joint model. Age had a smaller but still meaningful effect. These findings align closely with the individual analyses above and reinforce that gender and socioeconomic status were the primary determinants of survival on the Titanic.

```python
acc = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_proba)
print(f"Accuracy : {acc:.3f}")
print(f"ROC-AUC  : {auc:.3f}")
print(classification_report(y_test, y_pred, target_names=['Did Not Survive', 'Survived']))

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Logistic Regression - Model Evaluation', fontsize=14, fontweight='bold')

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['Did Not Survive', 'Survived'],
            yticklabels=['Did Not Survive', 'Survived'])
axes[0].set(title='Confusion Matrix', xlabel='Predicted', ylabel='True')

# ROC curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
axes[1].plot(fpr, tpr, lw=2, color='steelblue', label=f'AUC = {auc:.3f}')
axes[1].plot([0, 1], [0, 1], 'k--', lw=1)
axes[1].set(title='ROC Curve', xlabel='False Positive Rate', ylabel='True Positive Rate')
axes[1].legend()

# Feature coefficients — hue='feature' fixes the palette deprecation warning
feature_labels = ['Passenger Class', 'Sex (Female=1)', 'Age', 'Log Fare', 'Family Size', 'Alone', 'Embarked']
coef_df = pd.DataFrame({'feature': feature_labels, 'coef': lr.coef_[0]}).sort_values('coef')
sns.barplot(data=coef_df, y='feature', x='coef', hue='feature', ax=axes[2], palette='coolwarm_r', legend=False)
axes[2].axvline(0, color='black', lw=0.8, ls='--')
axes[2].set(title='Feature Coefficients (standardized)', xlabel='Coefficient', ylabel='')

plt.tight_layout()
plt.show()
```
